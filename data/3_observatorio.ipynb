```python
# Lista de URLs que contienen archivos Excel
urls = [
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/Amenazas_5.xlsx",
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/Amenazas2024_0.xlsx",
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/amenzas..xlsx",
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/amenazas_13.xls",
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/amenazas_11.xls",
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/amenazas_2020.xlsx",
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/amenazas_2019_1.xlsx",
    "https://www.policia.gov.co/sites/default/files/delitos-impacto/amenazas_2018_1.xlsx"
]

# Diccionario para almacenar los DataFrames
dataframes = {}

# Función para cargar un archivo Excel desde una URL
def cargar_excel_desde_url(url):
    try:
        # Enviar una solicitud GET para obtener el archivo
        respuesta = requests.get(url)

        # Verificar si la solicitud fue exitosa
        respuesta.raise_for_status()

        # Leer el archivo Excel en un DataFrame de pandas, usando la fila 10 como encabezados
        datos_excel = io.BytesIO(respuesta.content)
        df = pd.read_excel(datos_excel, header=10)  #  nombres de columnas

        # Eliminar las últimas cinco filas del DataFrame
        if len(df) >= 5:
            df = df.iloc[:-5]
        else:
            print(f"Advertencia: El DataFrame de {url} tiene menos de 5 filas, no se eliminaron filas.")

        return df
    except requests.exceptions.RequestException as e:
        print(f"Error al obtener la URL {url}: {e}")
        return None
    except pd.errors.ParserError as e:
        print(f"Error al analizar el archivo Excel de {url}: {e}")
        return None
    except Exception as e:
        print(f"Error inesperado con {url}: {e}")
        return None

# Iterar sobre las URLs y cargar cada archivo Excel
for i, url in enumerate(urls, 1):
    print(f"Cargando archivo Excel desde {url}")
    df = cargar_excel_desde_url(url)
    if df is not None:
        dataframes[f'df_{i}'] = df
        print(f"Cargado exitosamente df_{i}_amenazas con dimensiones {df.shape}")
    else:
        print(f"No se pudo cargar el archivo Excel desde {url}")

# Ejemplo: Mostrar las primeras filas de cada DataFrame
for nombre, df in dataframes.items():
    print(f"\nVista previa de {nombre}:")
    print(df.head())
```

La anterior forma de descarge evidencia que los encabezados no tienen un configuración homogénea.
