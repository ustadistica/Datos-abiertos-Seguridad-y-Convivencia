## Consolidación de la base de todos los delitos 2025

```python
# Cambiar nombre de una sola columna en las bases que emplean *
df_amenazas = df_amenazas.rename(columns={'*AGRUPA EDAD PERSONA*': 'AGRUPA_EDAD_PERSONA'})
df_delitos_sexuales = df_delitos_sexuales.rename(columns={'*AGRUPA EDAD PERSONA*': 'AGRUPA_EDAD_PERSONA'})
df_extorsion = df_extorsion.rename(columns={'*AGRUPA EDAD PERSONA*': 'AGRUPA_EDAD_PERSONA'})
df_homicidio_intencional = df_homicidio_intencional.rename(columns={'*AGRUPA EDAD PERSONA*': 'AGRUPA_EDAD_PERSONA'})
df_hurto_ganado = df_hurto_ganado.rename(columns={'*AGRUPA EDAD PERSONA*': 'AGRUPA_EDAD_PERSONA'})
```

**Reasoning**:
The previous code failed because the `delitos_por_municipio` DataFrame does not contain the 'Latitud' and 'longitud' columns. These columns are in the `df_delitos_divipola` DataFrame. I need to merge the `delitos_por_municipio` DataFrame with `df_delitos_divipola` to access the geographical coordinates.



```python
# Merge the delitos_por_municipio DataFrame with df_delitos_divipola to get geographical coordinates
delitos_por_municipio_geo = pd.merge(delitos_por_municipio, df_delitos_divipola[['MUNICIPIO', 'longitud', 'Latitud']], on='MUNICIPIO', how='left')

# Drop duplicate rows based on 'MUNICIPIO' to ensure unique municipality entries with coordinates
delitos_por_municipio_geo = delitos_por_municipio_geo.drop_duplicates(subset=['MUNICIPIO'])

# Create a base map centered on Colombia
m = folium.Map(location=[4.5709, -74.2973], zoom_start=6)
folium.TileLayer('cartodbdarkmatter').add_to(m)

# Add circle markers to the map, representing municipalities with the size and color of the markers based on the total number of crimes.
for index, row in delitos_por_municipio_geo.iterrows():
    folium.CircleMarker(
        location=[row['Latitud'], row['longitud']],
        radius=row['Total_Delitos'] / 5000, # Adjust radius for better visualization
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.6,
        tooltip=f"{row['MUNICIPIO']}: {row['Total_Delitos']} delitos"
    ).add_to(m)

m
```
