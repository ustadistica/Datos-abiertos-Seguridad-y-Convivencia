# Diseño de dimensiones





El primer paso es inspeccionar las columnas, los tipos de datos y los datos de muestra del marco de datos para identificar posibles dimensiones para el esquema en estrella. Hecho en la sección anterior.



## Creación de las tablas de dimensiones


Para cada dimensión identificada, se crea un nuevo marco de datos que contenga valores únicos y una clave principal generada.


```python
dim_arma_medio = df[['ARMAS_MEDIOS']].drop_duplicates().reset_index(drop=True)
dim_arma_medio['id_arma_medio'] = dim_arma_medio.index

dim_tipo_delito = df[['TIPO_DELITO']].drop_duplicates().reset_index(drop=True)
dim_tipo_delito['id_tipo_delito'] = dim_tipo_delito.index

dim_genero = df[['GENERO']].drop_duplicates().reset_index(drop=True)
dim_genero['id_genero'] = dim_genero.index

dim_agrupa_edad_persona = df[['AGRUPA_EDAD_PERSONA']].drop_duplicates().reset_index(drop=True)
dim_agrupa_edad_persona['id_agrupa_edad_persona'] = dim_agrupa_edad_persona.index

dim_delito = df[['DELITO']].drop_duplicates().reset_index(drop=True)
dim_delito['id_delito'] = dim_delito.index

dim_ubicacion = df[['Código Departamento', 'Nombre Departamento', 'Código Municipio', 'Nombre Municipio', 'Tipo: Municipio / Isla / Área no municipalizada']].drop_duplicates().reset_index(drop=True)
dim_ubicacion['id_ubicacion'] = dim_ubicacion.index

dim_año = df[['AÑO']].drop_duplicates().reset_index(drop=True)
dim_año['id_año'] = dim_año.index

dim_area_geografica = df[['ÁREA GEOGRÁFICA']].drop_duplicates().reset_index(drop=True)
dim_area_geografica['id_area_geografica'] = dim_area_geografica.index

df['FECHA'] = pd.to_datetime(df['FECHA'], errors='coerce')
dim_fecha = df[['FECHA']].drop_duplicates().dropna().reset_index(drop=True)
dim_fecha['id_fecha'] = dim_fecha.index

display(dim_arma_medio.head())
display(dim_tipo_delito.head())
display(dim_genero.head())
display(dim_agrupa_edad_persona.head())
display(dim_delito.head())
display(dim_ubicacion.head())
display(dim_año.head())
display(dim_area_geografica.head())
display(dim_fecha.head())
```

## Tabla de hechos (fact table)


Se identifican las columnas de hechos y se crea un nuevo marco de datos para la tabla de hechos, reemplazando las columnas de dimensión originales con las nuevas claves principales de las tablas de dimensión.

Se crea la tabla de hechos seleccionando las columnas especificadas y luego se combinan con cada tabla de dimensión para reemplazar las columnas de dimensión originales con sus respectivas claves principales.


```python
fact_table = df[['CANTIDAD', 'longitud', 'Latitud', 'TOTAL', 'ARMAS_MEDIOS', 'TIPO_DELITO', 'GENERO', 'AGRUPA_EDAD_PERSONA', 'DELITO', 'Código Departamento', 'Nombre Departamento', 'Código Municipio', 'Nombre Municipio', 'Tipo: Municipio / Isla / Área no municipalizada', 'AÑO', 'ÁREA GEOGRÁFICA', 'FECHA']].copy()

fact_table = pd.merge(fact_table, dim_arma_medio, on='ARMAS_MEDIOS', how='left')
fact_table = fact_table.drop('ARMAS_MEDIOS', axis=1)

fact_table = pd.merge(fact_table, dim_tipo_delito, on='TIPO_DELITO', how='left')
fact_table = fact_table.drop('TIPO_DELITO', axis=1)

fact_table = pd.merge(fact_table, dim_genero, on='GENERO', how='left')
fact_table = fact_table.drop('GENERO', axis=1)

fact_table = pd.merge(fact_table, dim_agrupa_edad_persona, on='AGRUPA_EDAD_PERSONA', how='left')
fact_table = fact_table.drop('AGRUPA_EDAD_PERSONA', axis=1)

fact_table = pd.merge(fact_table, dim_delito, on='DELITO', how='left')
fact_table = fact_table.drop('DELITO', axis=1)

fact_table = pd.merge(fact_table, dim_ubicacion, on=['Código Departamento', 'Nombre Departamento', 'Código Municipio', 'Nombre Municipio', 'Tipo: Municipio / Isla / Área no municipalizada'], how='left')
fact_table = fact_table.drop(['Código Departamento', 'Nombre Departamento', 'Código Municipio', 'Nombre Municipio', 'Tipo: Municipio / Isla / Área no municipalizada'], axis=1)

fact_table = pd.merge(fact_table, dim_año, on='AÑO', how='left')
fact_table = fact_table.drop('AÑO', axis=1)

fact_table = pd.merge(fact_table, dim_area_geografica, on='ÁREA GEOGRÁFICA', how='left')
fact_table = fact_table.drop('ÁREA GEOGRÁFICA', axis=1)

fact_table = pd.merge(fact_table, dim_fecha, on='FECHA', how='left')
fact_table = fact_table.drop('FECHA', axis=1)

display(fact_table.head())
```

* Las posibles columnas de hechos identificadas incluyen "CANTIDAD", "Longitud", "Latitud" y "TOTAL".

## Análisis de hechos por dimensiones

Análisis uniendo la tabla de hechos con cada tabla de dimensiones para comprender cómo se relacionan los hechos con las diferentes dimensiones. Esto incluirá agregaciones y visualizaciones.

Fusionar la tabla de hechos con cada tabla de dimensiones, agrupar por el atributo de dimensión y calcular la suma de "CANTIDAD" y el número de ocurrencias. Además, gestionar los valores faltantes en los atributos de dimensión rellenándolos con "Desconocido" antes de agrupar.

```python
# Handle missing values in dimension attributes by filling with 'Unknown'
dim_arma_medio['ARMAS_MEDIOS'] = dim_arma_medio['ARMAS_MEDIOS'].fillna('Unknown').replace(['NO REPORTA', 'NO REPORTADO', 'NO RESPORTADO', '-', 'NO REPORTADO '], 'Unknown')
dim_tipo_delito['TIPO_DELITO'] = dim_tipo_delito['TIPO_DELITO'].fillna('Unknown')
dim_genero['GENERO'] = dim_genero['GENERO'].fillna('Unknown').replace(['NO REPORTA', 'NO RESPORTADO', 'NO REPORTADO', 'NO REPORTADO ', '-'], 'Unknown')
dim_agrupa_edad_persona['AGRUPA_EDAD_PERSONA'] = dim_agrupa_edad_persona['AGRUPA_EDAD_PERSONA'].fillna('Unknown').replace(['NO REPORTA', 'NO RESPORTADO', 'NO REPORTADO', '-'], 'Unknown')
dim_delito['DELITO'] = dim_delito['DELITO'].fillna('Unknown')
dim_ubicacion['Nombre Departamento'] = dim_ubicacion['Nombre Departamento'].fillna('Unknown')
dim_ubicacion['Nombre Municipio'] = dim_ubicacion['Nombre Municipio'].fillna('Unknown')
dim_ubicacion['Tipo: Municipio / Isla / Área no municipalizada'] = dim_ubicacion['Tipo: Municipio / Isla / Área no municipalizada'].fillna('Unknown')
dim_año['AÑO'] = dim_año['AÑO'].fillna('Unknown')
dim_area_geografica['ÁREA GEOGRÁFICA'] = dim_area_geografica['ÁREA GEOGRÁFICA'].fillna('Unknown')

# Merge and aggregate for each dimension
print("--- Aggregation by ARMAS_MEDIOS ---")
merged_arma_medio = pd.merge(fact_table, dim_arma_medio, on='id_arma_medio', how='left')
agg_arma_medio = merged_arma_medio.groupby('ARMAS_MEDIOS').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_arma_medio.head())

print("\n--- Aggregation by TIPO_DELITO ---")
merged_tipo_delito = pd.merge(fact_table, dim_tipo_delito, on='id_tipo_delito', how='left')
agg_tipo_delito = merged_tipo_delito.groupby('TIPO_DELITO').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_tipo_delito.head())

print("\n--- Aggregation by GENERO ---")
merged_genero = pd.merge(fact_table, dim_genero, on='id_genero', how='left')
agg_genero = merged_genero.groupby('GENERO').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_genero.head())

print("\n--- Aggregation by AGRUPA_EDAD_PERSONA ---")
merged_agrupa_edad_persona = pd.merge(fact_table, dim_agrupa_edad_persona, on='id_agrupa_edad_persona', how='left')
agg_agrupa_edad_persona = merged_agrupa_edad_persona.groupby('AGRUPA_EDAD_PERSONA').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_agrupa_edad_persona.head())

print("\n--- Aggregation by DELITO ---")
merged_delito = pd.merge(fact_table, dim_delito, on='id_delito', how='left')
agg_delito = merged_delito.groupby('DELITO').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_delito.head())

print("\n--- Aggregation by Ubicacion (Departamento and Municipio) ---")
merged_ubicacion = pd.merge(fact_table, dim_ubicacion, on='id_ubicacion', how='left')
agg_ubicacion_dept_muni = merged_ubicacion.groupby(['Nombre Departamento', 'Nombre Municipio']).agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_ubicacion_dept_muni.head())

print("\n--- Aggregation by AÑO ---")
merged_año = pd.merge(fact_table, dim_año, on='id_año', how='left')
agg_año = merged_año.groupby('AÑO').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_año.head())

print("\n--- Aggregation by ÁREA GEOGRÁFICA ---")
merged_area_geografica = pd.merge(fact_table, dim_area_geografica, on='id_area_geografica', how='left')
agg_area_geografica = merged_area_geografica.groupby('ÁREA GEOGRÁFICA').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='total_cantidad', ascending=False)
display(agg_area_geografica.head())

print("\n--- Aggregation by FECHA (handling missing dates) ---")
# Filter out rows with missing date IDs before merging for date aggregation
fact_table_fecha_filtered = fact_table.dropna(subset=['id_fecha']).copy()
# Convert id_fecha to integer type after dropping NaNs
fact_table_fecha_filtered['id_fecha'] = fact_table_fecha_filtered['id_fecha'].astype(int)
merged_fecha = pd.merge(fact_table_fecha_filtered, dim_fecha, on='id_fecha', how='left')
agg_fecha = merged_fecha.groupby('FECHA').agg(
    total_cantidad=('CANTIDAD', 'sum'),
    count=('CANTIDAD', 'size')
).sort_values(by='FECHA') # Sort by date for time series analysis
display(agg_fecha.head())
display(agg_fecha.tail())
```

## Análisis de las relaciones entre dimensiones


Exploración de las relaciones entre dimensiones seleccionadas fusionándolas con la tabla de hechos y realizando agregaciones para comprender la distribución conjunta de hechos en estas dimensiones y visualización de las relaciones utilizando gráficos apropiados.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Explore relationship between Nombre Departamento and TIPO_DELITO
print("--- Relationship between Nombre Departamento and TIPO_DELITO ---")
# Merge fact table with relevant dimensions
merged_dept_delito = pd.merge(fact_table, dim_ubicacion[['id_ubicacion', 'Nombre Departamento']], on='id_ubicacion', how='left')
merged_dept_delito = pd.merge(merged_dept_delito, dim_tipo_delito[['id_tipo_delito', 'TIPO_DELITO']], on='id_tipo_delito', how='left')

# Group by department and type of crime and sum the quantity
agg_dept_delito = merged_dept_delito.groupby(['Nombre Departamento', 'TIPO_DELITO']).agg(
    total_cantidad=('CANTIDAD', 'sum')
).reset_index()

# Pivot the table for heatmap
pivot_dept_delito = agg_dept_delito.pivot_table(index='Nombre Departamento', columns='TIPO_DELITO', values='total_cantidad', fill_value=0)

# Select top N departments and crime types for better visualization
top_n_departments = agg_dept_delito.groupby('Nombre Departamento')['total_cantidad'].sum().nlargest(10).index
top_n_delitos = agg_dept_delito.groupby('TIPO_DELITO')['total_cantidad'].sum().nlargest(10).index

pivot_dept_delito_top = pivot_dept_delito.loc[top_n_departments, top_n_delitos]

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_dept_delito_top, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=.5)
plt.title('Total CANTIDAD by Top 10 Nombre Departamento and Top 10 TIPO_DELITO')
plt.xlabel('TIPO_DELITO')
plt.ylabel('Nombre Departamento')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Explore relationship between AGRUPA_EDAD_PERSONA and GENERO
print("\n--- Relationship between AGRUPA_EDAD_PERSONA and GENERO ---")
# Merge fact table with relevant dimensions
merged_edad_genero = pd.merge(fact_table, dim_agrupa_edad_persona[['id_agrupa_edad_persona', 'AGRUPA_EDAD_PERSONA']], on='id_agrupa_edad_persona', how='left')
merged_edad_genero = pd.merge(merged_edad_genero, dim_genero[['id_genero', 'GENERO']], on='id_genero', how='left')

# Group by age group and gender and sum the quantity
agg_edad_genero = merged_edad_genero.groupby(['AGRUPA_EDAD_PERSONA', 'GENERO']).agg(
    total_cantidad=('CANTIDAD', 'sum')
).reset_index()

# Pivot the table for heatmap
pivot_edad_genero = agg_edad_genero.pivot_table(index='AGRUPA_EDAD_PERSONA', columns='GENERO', values='total_cantidad', fill_value=0)

plt.figure(figsize=(8, 6))
sns.heatmap(pivot_edad_genero, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=.5)
plt.title('Total CANTIDAD by AGRUPA_EDAD_PERSONA and GENERO')
plt.xlabel('GENERO')
plt.ylabel('AGRUPA_EDAD_PERSONA')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Explore relationship between AÑO and TIPO_DELITO
print("\n--- Relationship between AÑO and TIPO_DELITO ---")
# Merge fact table with relevant dimensions
merged_año_delito = pd.merge(fact_table, dim_año[['id_año', 'AÑO']], on='id_año', how='left')
merged_año_delito = pd.merge(merged_año_delito, dim_tipo_delito[['id_tipo_delito', 'TIPO_DELITO']], on='id_tipo_delito', how='left')

# Group by year and type of crime and sum the quantity
agg_año_delito = merged_año_delito.groupby(['AÑO', 'TIPO_DELITO']).agg(
    total_cantidad=('CANTIDAD', 'sum')
).reset_index()

# Pivot the table for heatmap
pivot_año_delito = agg_año_delito.pivot_table(index='AÑO', columns='TIPO_DELITO', values='total_cantidad', fill_value=0)

# Select top N crime types for better visualization
top_n_delitos_año = agg_año_delito.groupby('TIPO_DELITO')['total_cantidad'].sum().nlargest(10).index

pivot_año_delito_top = pivot_año_delito.loc[:, top_n_delitos_año]

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_año_delito_top, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=".5")
plt.title('Total CANTIDAD by AÑO and Top 10 TIPO_DELITO')
plt.xlabel('TIPO_DELITO')
plt.ylabel('AÑO')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
