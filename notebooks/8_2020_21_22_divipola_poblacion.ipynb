### Unificación de las tablas

```python
# Check column names in each dataframe
print("Columns in df_2020:", df_2020.columns)
print("Columns in df_2021:", df_2021.columns)
print("Columns in df_2022:", df_2022.columns)
```

```python
# Rename columns in df_2020 to match df_2021 and df_2022
df_2020.rename(columns={'ARMAS MEDIOS': 'ARMAS_MEDIOS', 'FECHA HECHO': 'FECHA', 'AGRUPA EDAD PERSONA': 'AGRUPA_EDAD_PERSONA', 'CODIGO DANE': 'CODIGO_DANE', 'DELITO ': 'DELITO'}, inplace=True)

# Concatenate the three dataframes
df_combined = pd.concat([df_2020, df_2021, df_2022], ignore_index=True)

# Display the first few rows of the combined dataframe
display(df_combined.head())

# Display information about the combined dataframe
display(df_combined.info())
```

```python
# Valores faltantes
print("Missing values in df_combined:")
display(df_combined.isnull().sum())
```

```python
# Remove "(CT)" from the 'MUNICIPIO' column
df_combined['MUNICIPIO'] = df_combined['MUNICIPIO'].str.replace(' (CT)', '', regex=False)

# Display the unique values in the 'MUNICIPIO' column to verify the change (optional)
# display(df_combined['MUNICIPIO'].unique())
```

```python
df_combined.sample(10)
```

```python
# Drop the 'CODIGO_DANE' and 'DEPARTAMENTO' columns
df_combined = df_combined.drop(['CODIGO_DANE', 'DEPARTAMENTO'], axis=1)

# Display the first few rows and info of the modified dataframe to verify the change
display(df_combined.head())
display(df_combined.info())
```

```python
# Valores faltantes
print("Missing values in df_combined:")
display(df_combined.isnull().sum())
```

# Base de población

```python
# Cargar una hoja específica de un archivo Excel
# Omitir la primera fila de registros (al ser integrada a la de nombres de
# de columna figura como nan)
df_poblacion = pd.read_excel('/content/drive/MyDrive/Consultoria/datos_abiertos_seguridad_convivencia/data/raw/PPED-AreaMun-2018-2042_VP.xlsx', sheet_name='PobMunicipalxÁrea', header=7)
# Omitir la primera fila de registros
df_poblacion = df_poblacion.iloc[1:].copy()
df_poblacion.head()
df_poblacion.info()
```

```python
# Valores faltantes
print("Missing values in df_poblacion:")
display(df_poblacion.isnull().sum())
```

```python
# Filter the dataframe to include only the years 2020, 2021, and 2022
df_poblacion_filtered = df_poblacion[df_poblacion['AÑO'].isin([2020.0, 2021.0, 2022.0])].copy()

# Filter for 'Total' in the 'ÁREA GEOGRÁFICA' column
df_poblacion_filtered_total = df_poblacion_filtered[df_poblacion_filtered['ÁREA GEOGRÁFICA'] == 'Total'].copy()

# Display the first few rows and info of the filtered dataframe to verify the change
display(df_poblacion_filtered_total.head())
display(df_poblacion_filtered_total.info())
```

```python
# Convert the 'DPMP' column to uppercase
df_poblacion_filtered_total['DPMP'] = df_poblacion_filtered_total['DPMP'].str.upper()

# Display the first few rows to verify the change
display(df_poblacion_filtered_total.head())
```

```python
df_poblacion_filtered_total.sample(10)
```

```python
# Merge df_merged with df_poblacion_filtered
df_merged_population = pd.merge(df_merged, df_poblacion_filtered_total, left_on='Nombre Municipio', right_on='DPMP', how='left')

# Display the first few rows and info of the merged dataframe
display(df_merged_population.head())
display(df_merged_population.info())
```

```python
df_merged_population.sample(10)
```

```python
# Valores falantse en la base de delitos de 2022
print("Missing values in df_merged_population:")
display(df_merged_population.isnull().sum())
```

```python
# Visualización de valores faltantes
import missingno as msno
msno.matrix(df_merged_population)
plt.title('Missing values in df_merged_population')
plt.show()
```

La variable ARMAS_MEDIOS explica los medios empleados para cada uno de los delitos tipificados.

```python
# Gráfica de los primeros diez tipos de medios
plt.figure(figsize=(12, 6))
sns.countplot(data=df_merged_population, y='ARMAS_MEDIOS', order=df_merged_population['ARMAS_MEDIOS'].value_counts().nlargest(10).index)
plt.title('Top 10 Distribution of ARMAS_MEDIOS')
plt.xlabel('Count')
plt.ylabel('ARMAS_MEDIOS')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_merged_population, y='GENERO', order=df_merged_population['GENERO'].value_counts().index)
plt.title('Distribution of Crimes by Gender')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_merged_population, y='Nombre Municipio', order=df_merged_population['Nombre Municipio'].value_counts().nlargest(10).index)
plt.title('Top 10 Distribution of Crimes by Municipality')
plt.xlabel('Count')
plt.ylabel('Nombre Municipio')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_merged_population, y='AGRUPA_EDAD_PERSONA', order=df_merged_population['AGRUPA_EDAD_PERSONA'].value_counts().index)
plt.title('Distribution of Crimes by Age Group')
plt.xlabel('Count')
plt.ylabel('Age Group')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_merged_population, y='Nombre Departamento', order=df_merged_population['Nombre Departamento'].value_counts().nlargest(10).index)
plt.title('Top 10 Distribution of Crimes by Department')
plt.xlabel('Count')
plt.ylabel('Nombre Departamento')
plt.show()
```
