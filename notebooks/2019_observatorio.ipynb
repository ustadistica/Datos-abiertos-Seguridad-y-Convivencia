# Base de Proyección de población

```python
import requests
import pandas as pd
import numpy as np

# URL de la proyección del DANE
url_poblacion = 'https://www.dane.gov.co/files/censo2018/proyecciones-de-poblacion/Municipal/PPED-AreaSexoEdadMun-2018-2042_VP.xlsx'

# Leer el archivo con header en la fila 8 (header=7)
df_pob_raw = pd.read_excel(url_poblacion, sheet_name='PobMunicipalxÁreaSexoEdad', header=7)

# Manejo del encabezado de dos filas
new_columns = []
for col, col_name_row8 in zip(df_pob_raw.columns, df_pob_raw.iloc[0]):
    if 'Unnamed:' in col:
        new_columns.append(col_name_row8)
    else:
        new_columns.append(col)

df_pob_raw.columns = new_columns
df_pob = df_pob_raw.iloc[1:].copy()

# Limpiar nombres de columnas
df_pob.columns = df_pob.columns.str.strip()

# Renombrar columnas principales
df_pob = df_pob.rename(columns={
    'DPMP': 'MUNICIPIO',
    'AÑO': 'AÑO_POBLACION',
    'TOTAL': 'POBLACION_TOTAL'
})

# Convertir municipio a mayúsculas
df_pob['MUNICIPIO'] = df_pob['MUNICIPIO'].astype(str).str.upper()

# ------------------------------------------------------------
# Filtrar solo 2019 y ÁREA GEOGRÁFICA = 'Total'
# ------------------------------------------------------------
df_pob_2019 = df_pob[
    (df_pob['AÑO_POBLACION'] == 2019) &
    (df_pob['ÁREA GEOGRÁFICA'] == 'Total')
].drop(columns=['ÁREA GEOGRÁFICA'])

# Mostrar resultados
display(df_pob_2019.head())
display(df_pob_2019.info())
```

```python
import pandas as pd

# ===============================
# 1. Normalizar nombres de municipios
# ===============================
def limpiar_muni(df, col):
    if col in df.columns:
        df[col] = df[col].astype(str).str.upper().str.replace(r"\(CT\)", "", regex=True).str.strip()
    return df

df_consolidado_2019 = limpiar_muni(df_consolidado_2019, "MUNICIPIO")
df_divipola = limpiar_muni(df_divipola, "nom_mpio")
df_pob_2019 = limpiar_muni(df_pob_2019, "MUNICIPIO")

# ===============================
# 2. Merge consolidado + divipola
# ===============================
df_merge1 = pd.merge(
    df_consolidado_2019,
    df_divipola[["cod_dpto", "dpto", "cod_mpio", "nom_mpio", "tipo_municipio", "latitud", "longitud"]],
    left_on="MUNICIPIO",
    right_on="nom_mpio",
    how="left"
)

# ===============================
# 3. Extraer año para unir con población
# ===============================
df_merge1["AÑO_HECHO"] = df_merge1["FECHA HECHO"].dt.year

df_master = pd.merge(
    df_merge1,
    df_pob_2019,  # contiene DP, DPNOM, MPIO, MUNICIPIO, AÑO_POBLACION, POBLACION_TOTAL, etc.
    left_on=["MUNICIPIO", "AÑO_HECHO"],
    right_on=["MUNICIPIO", "AÑO_POBLACION"],
    how="left"
)

# ===============================
# 4. Crear columnas finales unificadas
# ===============================

# Departamento (nombre) -> tomamos directamente el de delitos
df_master = df_master.rename(columns={"DEPARTAMENTO": "DEPTO_FINAL", "MUNICIPIO": "MUNICIPIO_FINAL"})

# Código de municipio combinado
df_master["COD_MPIO_FINAL"] = (
    df_master["CODIGO DANE"]
    .combine_first(df_master.get("cod_mpio"))
    .combine_first(df_master.get("MPIO"))
)

# Código de departamento combinado
df_master["COD_DPTO_FINAL"] = (
    df_master.get("cod_dpto")
    .combine_first(df_master.get("DP"))
)

# ===============================
# 5. Eliminar columnas duplicadas innecesarias
# ===============================
cols_to_drop = [
    "nom_mpio", "cod_mpio", "MPIO", "DP", "DPNOM", "dpto",
    "CODIGO DANE", "AÑO_POBLACION"
]
df_master = df_master.drop(columns=[c for c in cols_to_drop if c in df_master.columns])

# ===============================
# 6. Ver resultado
# ===============================
print("Columnas finales:", df_master.columns.tolist())
display(df_master.head(10))
```

```python
# ===============================
# 7. Limpieza y reordenamiento de códigos
# ===============================

# Convertir a enteros limpios (si hay nulos se rellenan con 0)
df_master["COD_MPIO_FINAL"] = (
    pd.to_numeric(df_master["COD_MPIO_FINAL"], errors="coerce")
    .fillna(0)
    .astype(int)
)
df_master["COD_DPTO_FINAL"] = (
    pd.to_numeric(df_master["COD_DPTO_FINAL"], errors="coerce")
    .fillna(0)
    .astype(int)
)

# Reordenar columnas → primero COD_MPIO_FINAL y COD_DPTO_FINAL
cols = list(df_master.columns)
cols.remove("COD_MPIO_FINAL")
cols.remove("COD_DPTO_FINAL")
ordered_cols = ["COD_MPIO_FINAL", "COD_DPTO_FINAL"] + cols
df_master = df_master[ordered_cols]

# ===============================
# 8. Ver resultado limpio
# ===============================
print("Columnas finales ordenadas:", df_master.columns[:10].tolist(), "...")
display(df_master.head(10))
```

```python
# ===============================
# Agrupación de delitos por departamento y tipo de delito
# ===============================

crime_by_department = (
    df_master
    .groupby(['DEPTO_FINAL', 'TIPO_DELITO'], as_index=False)['CANTIDAD']
    .sum()
)

# Pasar de Series a DataFrame con unstack/pivot
df_combinado_final_2019 = crime_by_department.pivot(
    index='DEPTO_FINAL',
    columns='TIPO_DELITO',
    values='CANTIDAD'
).fillna(0)

# ===============================
# Ver resultados
# ===============================
print("Tabla de delitos por departamento y tipo de delito:")
display(df_combinado_final_2019.head(10))

```

### Análisis Bivariado para Delitos por Departamento (df_combinado_final_2019)

#### Num vs Num (Correlación entre Tipos de Delitos)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlación (Pearson por defecto)
correlation_matrix = df_combinado_final_2019.corr()

print("Matriz de Correlación de Pearson entre Tipos de Delitos:")
display(correlation_matrix)

# Visualizar la matriz de correlación con un heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlación entre Tipos de Delitos por Departamento')
plt.show()
```
