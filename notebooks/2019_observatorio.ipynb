## Proponer carga de datos estandarizada

Para abordar la inconsistencia en los encabezados y pies de página, se puede implementar un enfoque más robusto que identifique dinámicamente las filas relevantes en cada archivo.

Una estrategia posible es:

1.  **Leer las primeras filas del archivo**: Cargar un número suficiente de filas iniciales del archivo (por ejemplo, las primeras 15 o 20 filas) en un DataFrame temporal.
2.  **Identificar la fila del encabezado**: Buscar en este DataFrame temporal una fila que contenga nombres de columnas esperados (como 'DEPARTAMENTO', 'MUNICIPIO', 'FECHA HECHO', etc.). Se puede hacer una verificación parcial o buscar un conjunto mínimo de columnas clave.
3.  **Identificar las filas del pie de página**: De manera similar, se pueden identificar las últimas filas que no contienen datos relevantes (como totales o notas al pie) buscando patrones o palabras clave ("TOTAL", "FUENTE", etc.) o basándose en el número de valores no nulos en las últimas filas.
4.  **Cargar el archivo completo con las filas correctas**: Una vez identificadas la fila del encabezado y las filas a omitir al final, cargar el archivo Excel completo utilizando los parámetros `header` y `skipfooter` de `pd.read_excel` con los valores determinados dinámicamente. Es importante usar el motor `engine='openpyxl'` para archivos `.xlsx`.
5.  **Manejar errores**: Implementar manejo de errores (`try-except`) en caso de que la estructura del archivo no coincida con lo esperado o si la URL no es accesible.

Este enfoque requiere un análisis previo de la posible variabilidad en la estructura de los archivos a lo largo de los años, pero una vez implementado, hará que el proceso de carga sea mucho más resiliente.

Se han identificado los siguientes riesgos e inconsistencias en el código de 2019:

*   **Diferencias en encabezados y pies de página**: Los archivos de origen tienen diferentes números de filas de encabezado y pie de página que se eliminan manualmente (`header=10`, `iloc[:-5]`, `header=9`, `iloc[:-3]`). Esto es inconsistente y propenso a errores si la estructura de los archivos cambia en el futuro.
*   **Columnas faltantes en algunas bases**: Algunas columnas como 'GENERO' y 'AGRUPA EDAD PERSONA' no existen en todos los archivos (por ejemplo, en `df_hurto_automotores2019` y `df_terrorismo2019`), lo que resulta en valores `NaN` en el dataframe consolidado.
*   **Nombres de columnas inconsistentes**: Aunque se ha intentado estandarizar la columna '*AGRUPA EDAD PERSONA*', otros nombres de columnas pueden variar o contener caracteres especiales o espacios que dificulten su manejo.
*   **Problemas de estandarización de municipios**: La columna 'MUNICIPIO' contiene inconsistencias como '(CT)' y la mezcla de mayúsculas y minúsculas, como se observa en los valores únicos. Aunque se ha realizado un intento de limpieza, es posible que queden variaciones que impidan un cruce correcto con datos externos como Divipola.
*   **Formatos de fecha**: Aunque `pd.to_datetime` se utiliza para convertir la columna 'FECHA HECHO', las inconsistencias en los formatos de fecha de origen podrían causar errores durante la conversión si no se manejan adecuadamente (se utilizó `errors='coerce'`, que convierte las fechas no válidas a `NaT`).
*   **Valores NaN**: Existen múltiples columnas con un número significativo de valores faltantes (`NaN`), como 'ARMAS MEDIOS', 'GENERO', 'AGRUPA EDAD PERSONA' y 'DESCRIPCION CONDUCTA'. La forma en que se manejen estos valores afectará los análisis posteriores.
*   **Comparaciones entre años**: Dado que el proceso de carga y limpieza es manual y específico para 2019, aplicar el mismo código directamente a datos de otros años con estructuras o columnas diferentes generará errores. Se necesita un proceso más generalizado.

### UNIÓN Y LIMPIEZA

```python
# Lista de dataframes a concatenar
dfs_2019 = [
    df_amenazas2019,
    df_extorsion2019,
    df_lesiones_transito2019,
    df_violencia_intra2019,
    df_secuestro2019,
    df_hurto_residencias2019,
    df_hurto_automotores2019,
    df_hurto_personas2019,
    df_hurto_motos2019,
    df_hurto_ganado2019,
    df_hurto_financiera2019,
    df_pirateria_terrestre2019,
    df_lesiones_personales2019,
    df_delitos_sexuales2019
]

# Concatenar los dataframes
df_consolidado_2019 = pd.concat(dfs_2019, ignore_index=True)

# Mostrar las primeras filas y la información del nuevo dataframe
display(df_consolidado_2019.head())
display(df_consolidado_2019.info())
```

```python
import numpy as np

# Replace '-' with NaN in the 'GENERO' column
df_consolidado_2019['GENERO'] = df_consolidado_2019['GENERO'].replace({'-': np.nan})

# Display unique values in 'GENERO' to verify the change
print("Valores únicos en la columna 'GENERO' después de reemplazar '-':")
display(df_consolidado_2019['GENERO'].unique())
```

```python
# Remove '(CT)' and convert to uppercase in the 'MUNICIPIO' column
df_consolidado_2019['MUNICIPIO'] = df_consolidado_2019['MUNICIPIO'].str.replace(' \(CT\)', '', regex=True).str.upper()

# Display unique values or a sample to verify the changes
print("Valores únicos en la columna 'MUNICIPIO' después de la limpieza:")
display(df_consolidado_2019['MUNICIPIO'].unique())
```

```python
# Mostrar 5 registros aleatorios del dataframe consolidado
display(df_consolidado_2019.sample(10))
```

```python
# Agrupar por tipo de delito y sumar la cantidad
delito_counts = df_consolidado_2019.groupby('TIPO_DELITO')['CANTIDAD'].sum().sort_values(ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
delito_counts.plot(kind='bar')
plt.title('Cantidad total de delitos por tipo en 2019')
plt.xlabel('Tipo de Delito')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Convertir la columna 'FECHA HECHO' a datetime objects
df_consolidado_2019['FECHA HECHO'] = pd.to_datetime(df_consolidado_2019['FECHA HECHO'], errors='coerce')

# Display the info of the dataframe to verify the change
display(df_consolidado_2019.info())
```
