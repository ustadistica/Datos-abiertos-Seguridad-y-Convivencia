```python
# Define a consistent set of columns to keep
columns_to_keep = [
    'ARMAS_MEDIOS', 'DEPARTAMENTO', 'MUNICIPIO', 'FECHA', 'GENERO',
    'AGRUPA_EDAD_PERSONA', 'CODIGO_DANE', 'CANTIDAD', 'TIPO_DELITO'
]

# Prepare each dataframe by ensuring consistent columns and cleaning names
def prepare_dataframe(df, columns_to_keep, name):
    # Create a copy to avoid modifying the original dataframe in place
    df_prepared = df.copy()

    # Add missing columns from columns_to_keep with None values
    for col in columns_to_keep:
        if col not in df_prepared.columns:
            df_prepared[col] = None

    # Clean column names
    df_prepared.columns = df_prepared.columns.str.strip()
    df_prepared.columns = df_prepared.columns.str.replace('[^A-Za-z0-9_]+', '_', regex=True)
    df_prepared.columns = df_prepared.columns.str.upper()

    # Rename specific columns with known variations (apply to cleaned names)
    rename_map = {
        'ARMAS_MEDIO': 'ARMAS_MEDIOS',
        'ARMA_EMPLEADA': 'ARMAS_MEDIOS',
        'CODIGO_DANE_': 'CODIGO_DANE',
        'MUNICICPIO': 'MUNICIPIO',
        'FECHA_HECHO': 'FECHA',
        'ARMA_MEDIOS': 'ARMAS_MEDIOS',
        'FECHA_HECHO_': 'FECHA',
        'CODIGO_DANE': 'CODIGO_DANE',
        'FECHA': 'FECHA',
        'CANTIDAD': 'CANTIDAD',
        'GENERO': 'GENERO',
        'AGRUPA_EDAD_PERSONA': 'AGRUPA_EDAD_PERSONA',
        'TIPO_DELITO': 'TIPO_DELITO'
    }
    df_prepared = df_prepared.rename(columns=rename_map)

    # Ensure only columns to keep are present and in the correct order
    # Filter columns that are in both the dataframe and the columns_to_keep list
    cols_to_select = [col for col in columns_to_keep if col in df_prepared.columns]

    # Return the dataframe with only the selected columns in the specified order
    return df_prepared[cols_to_select]


# Prepare 2021 dataframes
df_hurto_personas2021_prepared = prepare_dataframe(df_hurto_personas2021.copy(), columns_to_keep, 'df_hurto_personas2021')
df_pirateria_terrestre2021_prepared = prepare_dataframe(df_pirateria_terrestre2021.copy(), columns_to_keep, 'df_pirateria_terrestre2021')
df_hurto_residencias2021_prepared = prepare_dataframe(df_hurto_residencias2021.copy(), columns_to_keep, 'df_hurto_residencias2021')
df_hurto_ganado2021_prepared = prepare_dataframe(df_hurto_ganado2021.copy(), columns_to_keep, 'df_hurto_ganado2021')
df_hurto_automotores2021_prepared = prepare_dataframe(df_hurto_automotores2021.copy(), columns_to_keep, 'df_hurto_automotores2021')
df_amenazas2021_prepared = prepare_dataframe(df_amenazas2021.copy(), columns_to_keep, 'df_amenazas2021')
df_hurto_financiera2021_prepared = prepare_dataframe(df_hurto_financiera2021.copy(), columns_to_keep, 'df_hurto_financiera2021')
df_delitos_sexuales2021_prepared = prepare_dataframe(df_delitos_sexuales2021.copy(), columns_to_keep, 'df_delitos_sexuales2021')
df_lesiones_transito2021_prepared = prepare_dataframe(df_lesiones_transito2021.copy(), columns_to_keep, 'df_lesiones_transito2021')
df_extorsion2021_prepared = prepare_dataframe(df_extorsion2021.copy(), columns_to_keep, 'df_extorsion2021')
df_hurto_motos2021_prepared = prepare_dataframe(df_hurto_motos2021.copy(), columns_to_keep, 'df_hurto_motos2021')
df_lesiones_personales2021_prepared = prepare_dataframe(df_lesiones_personales2021.copy(), columns_to_keep, 'df_lesiones_personales2021')
df_homicidios_transito2021_prepared = prepare_dataframe(df_homicidios_transito2021.copy(), columns_to_keep, 'df_homicidios_transito2021')
df_secuestro2021_prepared = prepare_dataframe(df_secuestro2021.copy(), columns_to_keep, 'df_secuestro2021')
df_homicidio_intencional2021_prepared = prepare_dataframe(df_homicidio_intencional2021.copy(), columns_to_keep, 'df_homicidio_intencional2021')


# Combine 2021 dataframes
df_2021 = pd.concat([
    df_hurto_personas2021_prepared,
    df_pirateria_terrestre2021_prepared,
    df_hurto_residencias2021_prepared,
    df_hurto_ganado2021_prepared,
    df_hurto_automotores2021_prepared,
    df_amenazas2021_prepared,
    df_hurto_financiera2021_prepared,
    df_delitos_sexuales2021_prepared,
    df_lesiones_transito2021_prepared,
    df_extorsion2021_prepared,
    df_hurto_motos2021_prepared,
    df_lesiones_personales2021_prepared,
    df_homicidios_transito2021_prepared,
    df_secuestro2021_prepared,
    df_homicidio_intencional2021_prepared
], ignore_index=True)


# Prepare each dataframe for 2022
df_hurto_residencias2022_prepared = prepare_dataframe(df_hurto_residencias2022.copy(), columns_to_keep, 'df_hurto_residencias2022')
df_homicidio_intencional2022_prepared = prepare_dataframe(df_homicidio_intencional2022.copy(), columns_to_keep, 'df_homicidio_intencional2022')
df_hurto_automotores2022_prepared = prepare_dataframe(df_hurto_automotores2022.copy(), columns_to_keep, 'df_hurto_automotores2022')
df_hurto_ganado2022_prepared = prepare_dataframe(df_hurto_ganado2022.copy(), columns_to_keep, 'df_hurto_ganado2022')
df_hurto_financiera2022_prepared = prepare_dataframe(df_hurto_financiera2022.copy(), columns_to_keep, 'df_hurto_financiera2022')
df_amenazas2022_prepared = prepare_dataframe(df_amenazas2022.copy(), columns_to_keep, 'df_amenazas2022')
df_lesiones_transito2022_prepared = prepare_dataframe(df_lesiones_transito2022.copy(), columns_to_keep, 'df_lesiones_transito2022')
df_delitos_sexuales2022_prepared = prepare_dataframe(df_delitos_sexuales2022.copy(), columns_to_keep, 'df_delitos_sexuales2022')
df_lesiones_personales2022_prepared = prepare_dataframe(df_lesiones_personales2022.copy(), columns_to_keep, 'df_lesiones_personales2022')
df_extorsion2022_prepared = prepare_dataframe(df_extorsion2022.copy(), columns_to_keep, 'df_extorsion2022')
df_secuestro2022_prepared = prepare_dataframe(df_secuestro2022.copy(), columns_to_keep, 'df_secuestro2022')
df_terrorismo2022_prepared = prepare_dataframe(df_terrorismo2022.copy(), columns_to_keep, 'df_terrorismo2022')
df_hurto_motos2022_prepared = prepare_dataframe(df_hurto_motos2022.copy(), columns_to_keep, 'df_hurto_motos2022')
df_hurto_personas2022_prepared = prepare_dataframe(df_hurto_personas2022.copy(), columns_to_keep, 'df_hurto_personas2022')
df_pirateria_terrestre2022_prepared = prepare_dataframe(df_pirateria_terrestre2022.copy(), columns_to_keep, 'df_pirateria_terrestre2022')

# Combine 2022 dataframes
df_2022 = pd.concat([
    df_hurto_residencias2022_prepared,
    df_homicidio_intencional2022_prepared,
    df_hurto_automotores2022_prepared,
    df_hurto_ganado2022_prepared,
    df_hurto_financiera2022_prepared,
    df_amenazas2022_prepared,
    df_lesiones_transito2022_prepared,
    df_delitos_sexuales2022_prepared,
    df_lesiones_personales2022_prepared,
    df_extorsion2022_prepared,
    df_secuestro2022_prepared,
    df_terrorismo2022_prepared,
    df_hurto_motos2022_prepared,
    df_hurto_personas2022_prepared,
    df_pirateria_terrestre2022_prepared
], ignore_index=True)


# Combine 2021 and 2022 dataframes
df_combined = pd.concat([df_2021, df_2022], ignore_index=True)

# Convert 'FECHA' to datetime, coercing errors
df_combined['FECHA'] = pd.to_datetime(df_combined['FECHA'], errors='coerce')

# Drop rows where 'FECHA' is NaT (due to coercion of invalid dates) or 'CANTIDAD' is NaN
df_combined.dropna(subset=['FECHA', 'CANTIDAD'], inplace=True)

# Convert 'CANTIDAD' to integer
df_combined['CANTIDAD'] = df_combined['CANTIDAD'].astype(int)

# Add 'AÑO' column
df_combined['AÑO'] = df_combined['FECHA'].dt.year

# Display the first few rows of the combined dataframe
print("df_2021 head:")
display(df_2021.head())

# Display information about the combined dataframe
print("\ndf_2021 info:")
display(df_2021.info())

print("\ndf_combined head:")
display(df_combined.head())

print("\ndf_combined info:")
display(df_combined.info())
```
