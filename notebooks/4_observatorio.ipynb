### UNIÓN Y LIMPIEZA

```python
# Lista de dataframes a concatenar
dfs_2024 = [
    df_amenazas2024,
    df_extorsion2024,
    df_lesiones_transito2024,
    df_violencia_intra2024,
    df_secuestro2024,
    df_hurto_residencias2024,
    df_hurto_automotores2024,
    df_hurto_personas2024,
    df_hurto_motos2024,
    df_hurto_ganado2024,
    df_hurto_financiera2024,
    df_pirateria_terrestre2024,
    df_lesiones_personales2024,
    df_delitos_sexuales2024
]

# Concatenar los dataframes
df_consolidado_2024 = pd.concat(dfs_2024, ignore_index=True)

# Mostrar las primeras filas y la información del nuevo dataframe
display(df_consolidado_2024.head())
display(df_consolidado_2024.info())
```

```python
import numpy as np

# Replace None with NaN in the 'GENERO' column
df_consolidado_2024['GENERO'] = df_consolidado_2024['GENERO'].replace({None: np.nan})

# Display unique values in 'GENERO' to verify the change
print("Valores únicos en la columna 'GENERO' después de la conversión:")
display(df_consolidado_2024['GENERO'].unique())
```

```python
import numpy as np

# Replace '-' with NaN in the 'GENERO' column
df_consolidado_2024['GENERO'] = df_consolidado_2024['GENERO'].replace({'-': np.nan})

# Display unique values in 'GENERO' to verify the change
print("Valores únicos en la columna 'GENERO' después de reemplazar '-':")
display(df_consolidado_2024['GENERO'].unique())
```

```python
# Remove '(CT)' and convert to uppercase in the 'MUNICIPIO' column
df_consolidado_2024['MUNICIPIO'] = df_consolidado_2024['MUNICIPIO'].str.replace(' \(CT\)', '', regex=True).str.upper()

# Display unique values or a sample to verify the changes
print("Valores únicos en la columna 'MUNICIPIO' después de la limpieza:")
display(df_consolidado_2024['MUNICIPIO'].unique())
```

```python
# Mostrar 5 registros aleatorios del dataframe consolidado
display(df_consolidado_2024.sample(10))
```

```python
# Agrupar por tipo de delito y sumar la cantidad
delito_counts = df_consolidado_2024.groupby('TIPO_DELITO')['CANTIDAD'].sum().sort_values(ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
delito_counts.plot(kind='bar')
plt.title('Cantidad total de delitos por tipo en 2024')
plt.xlabel('Tipo de Delito')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Convertir la columna 'FECHA HECHO' a datetime objects
df_consolidado_2024['FECHA HECHO'] = pd.to_datetime(df_consolidado_2024['FECHA HECHO'], errors='coerce')

# Display the info of the dataframe to verify the change
display(df_consolidado_2024.info())
```

```python
# Define a renaming map for cleaning leading/trailing spaces
cleaning_map = {col: col.strip() for name, df in dfs_2023_dict.items() for col in df.columns}

# Define a renaming map for standardizing variations
standardizing_map = {
    'ARMA MEDIO': 'ARMAS MEDIOS',
    'ARMAS_MEDIOS': 'ARMAS MEDIOS',
    'FECHA': 'FECHA HECHO',
    'CODIGO_DANE': 'CODIGO DANE',
    '*AGRUPA_EDAD_PERSONA': 'AGRUPA_EDAD_PERSONA'
}

# Apply renaming to each dataframe in the dictionary
for name, df in dfs_2023_dict.items():
    # Apply cleaning map first
    df = df.rename(columns=cleaning_map)
    # Only rename columns that exist in the dataframe for standardizing map
    cols_to_standardize = {k: v for k, v in standardizing_map.items() if k in df.columns}
    dfs_2023_dict[name] = df.rename(columns=cols_to_standardize)

# Print column names again to verify the changes
for name, df in dfs_2023_dict.items():
    print(f"Cleaned column names for {name}:")
    display(df.columns)
    print("-" * 30)
```

```python
# Concatenate the dataframes vertically now that the column names are standardized
df_consolidado_2023 = pd.concat(dfs_2023_dict.values(), ignore_index=True)

# Display the first rows and information of the new dataframe
display(df_consolidado_2023.head())
display(df_consolidado_2023.info())
```

```python
import numpy as np

# Convert the 'FECHA HECHO' column to datetime objects
df_consolidado_2023['FECHA HECHO'] = pd.to_datetime(df_consolidado_2023['FECHA HECHO'], errors='coerce')

# Replace '-' and 'NO REPORTADO' with NaN in the 'GENERO' column
df_consolidado_2023['GENERO'] = df_consolidado_2023['GENERO'].replace({'-': np.nan, 'NO REPORTADO': np.nan})

# Remove '(CT)' and convert to uppercase in the 'MUNICIPIO' column
df_consolidado_2023['MUNICIPIO'] = df_consolidado_2023['MUNICIPIO'].str.replace(r' \(CT\)', '', regex=True).str.upper()

# Calculate the quantity of missing values per column
missing_values_2023 = df_consolidado_2023.isnull().sum()

# Display the table of missing values
display(missing_values_2023)
```
